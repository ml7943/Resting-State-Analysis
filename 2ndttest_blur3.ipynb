```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from scipy.stats import pearsonr

# Calculate correlations between TEPSA and other variables
corr_variables = ['amyg', 'antPFC2', 'dlPFC2', 'IFC', 'cg_3', 'cg_25_2', 'dACC2', 'DCaud', 'dmPFC3', 'Hb_mni', 'hipp_thr', 'rl_accumbens', 'sgACC5', 'vmPFC2']
correlation_results = {}

for var in corr_variables:
    r, p_value = pearsonr(df['TEPSA'], df[var])
    correlation_results[var] = {'correlation_coefficient': r, 'p_value': p_value}

# Perform partial correlation analysis controlling for covariates
partial_corr_results = {}

for var in corr_variables:
    covariates = ['SUB_gender_1_F_0_M', 'SUB_Age', 'Medication_1_yes_0_no', 'Daw']
    X = sm.add_constant(df[[var] + covariates])
    partial_corr_model = sm.OLS(df['TEPSA'], X).fit()
    partial_corr_results[var] = {'coefficient': partial_corr_model.params[var],
                                 'p_value': partial_corr_model.pvalues[var],
                                 'effect_size': partial_corr_model.rsquared}

# Print the results
for var in corr_variables:
    print(f"Correlation between TEPSA and {var}:")
    print(f"Pearson correlation coefficient: {correlation_results[var]['correlation_coefficient']}")
    print(f"Pearson correlation p-value: {correlation_results[var]['p_value']}")
    print(f"Partial correlation coefficient: {partial_corr_results[var]['coefficient']}")
    print(f"Partial correlation p-value: {partial_corr_results[var]['p_value']}")
    print(f"Partial correlation effect size (R-squared): {partial_corr_results[var]['effect_size']}")
    print("\n")
```
